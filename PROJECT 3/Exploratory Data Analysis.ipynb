[Pelajari nilai yang hilang untuk setiap kolom secara lebih detail dan tentukan apa yang harus dilakukan dengan nilai tersebut berdasarkan hipotesis.]

Nilai yang hilang mungkin dikarenakan kurangnya input pada beberapa bagian kolom.Adapun juga terdapat kolom `odometer` salah isi dimana terdapat nilai `0`. Langkah langkah yang harus dilakukan yaitu membuat fungsi dengan mencari nilai median tiap data dengan kolom yang sesuai. Lalu nilai tersebut diisi pada nilai yang hilang. 

````python
#fungsi untuk mengisi nilai hilang
def fillna_group (dataframe, col_missing, col_group):
    median = pd.pivot_table(dataframe, columns=col_group, values=col_missing, aggfunc='median')

    for column in median.columns:
        dataframe.loc[(dataframe[col_group]==column) & (dataframe[col_missing].isnull()), col_missing] = median[column][0]

    return dataframe
````

````python
#mengisi nilai hilang model_year
df = fillna_group(df, 'model_year', 'model')
````

````python
#mengisi nilai hilang odometer
df = fillna_group(df , 'odometer','condition')
````

````python
#mengisi nilai hilang model_year
df = fillna_group(df, 'cylinders', 'model')
````

````python
#mengisi nilai hilang is_4wd
df['is_4wd'] = df['is_4wd'].fillna(0)
````

````python
##mengisi nilai hilang paint_color
df['paint_color'].fillna(df['paint_color'].mode()[0], inplace=True)
````

````python
df.isnull().sum()
````

Nilai yang hilang telah diisi oleh hasil median dari masing masing kolom. Lalu untuk paint_color diisi menggunakan modus, dikarenakan value nilai hilang tersebut bukan berbentuk nilai akan tetapi berupa karakter. Nilai hilang sudah teratasi.

Selain itu nilai `condition` dan `cylinder` agak terlihat tidak mencurigakan tetapi outlier pada kolom tersebut perlu diperiksa. Selanjutnya yaotu menangani outlier dari semua kolom tersebut

````python
# 2. Bikin function untuk hapus data outlier
def upper_lower_whisker(column_data):
        q1 = column_data.quantile(0.25)
        q3 = column_data.quantile(0.75)
        iqr = q3 - q1
        lower_whisker = q1 - 1.5 * iqr
        upper_whisker = q3 + 1.5 * iqr
        return lower_whisker, upper_whisker
````

````python
def outlier_cleansing(dataframe, column):

    lower, upper = upper_lower_whisker(dataframe[column])
    print(f'lower values : {lower}')
    print(f'upper values : {upper}')

    plt.title('Before cleaning outlier!')
    sns.boxplot(data=dataframe, x=column)
    plt.show()

    dataframe = dataframe[(dataframe[column] > lower) & (dataframe[column] < upper)]

    lower_check, upper_check = upper_lower_whisker(dataframe[column])
    print(f'lower_check values : {lower_check}')
    print(f'upper_check values : {upper_check}')

    plt.title('After cleaning outlier!')
    sns.boxplot(data=dataframe, x=column)
    plt.show()

    if lower == lower_check or upper == upper_check:
        print(f'\n====== [FINISHED] outlier in [{column}] column has been clear! ====== \n')
        return dataframe

    else:
        print(f'\n------ [RE-START] cleaning outlier in {column} column ------ \n')
        dataframe = outlier_cleansing(dataframe, column)
        return dataframe
````
