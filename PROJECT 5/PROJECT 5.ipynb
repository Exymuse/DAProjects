```python
# Fungsi untuk mengisi nilai yang hilang di year_of_release berdasarkan name
def fill_missing_year(row):
    return df.loc[df['name'] == row['name'], 'year_of_release'].max() if pd.isnull(row['year_of_release']) else row['year_of_release']

# Isi nilai yang hilang berdasarkan tahun rilis yang sudah diketahui
df['year_of_release'] = df.apply(fill_missing_year, axis=1) 
```

## 3.4 Dinamika Penjualan Game pada platform <a id='game_dynamic'></a>


### 3.4.1 Distribusi  Penjualan Game berdasarkan Platform per Tahun <a id='game_platform_dist'></a>


```python
# Keseluruhan data penjualan diatas 1995
all_data = df[df['year_of_release'] >= 1995]
platforms = all_data['platform'].unique()
```

```python
data = []  # List untuk menyimpan data

for platform in platforms:
    platform_data = all_data[all_data['platform'] == platform]
    sales_per_year = platform_data.groupby('year_of_release')['total_sales'].sum()
    max_sales_year = sales_per_year.idxmax()
    
    if max_sales_year is not None:
        max_sales_value = sales_per_year.max()
        data.append({
            'Platform': platform,
            'Tahun Penjualan Tertinggi': max_sales_year,
            'Total Penjualan': max_sales_value
        })

# Buat DataFrame
result_df = pd.DataFrame(data)

# Urutkan berdasarkan total penjualan
result_df = result_df.sort_values(by='Total Penjualan', ascending=False)


# Tampilkan sebagai tabel
result_df
```

```python
def sales_per_year(df):

    # Pivot the data to have years as columns and platforms as rows
    sales_per_year_pivot = df.pivot_table(index='platform', columns='year_of_release', 
                                          values='total_sales', aggfunc='sum')
    
    # Sort the data by total sales in descending order
    sales_per_year_pivot['total_sales'] = sales_per_year_pivot.sum(axis=1)
    sales_per_year_pivot = sales_per_year_pivot.sort_values(by='total_sales', ascending=False)
    
    
    num_subplots = len(sales_per_year_pivot)
    rows = (num_subplots + 2) // 3  # Round up to the nearest multiple of 3

    fig, axs = plt.subplots(rows, 3, figsize=(18, rows * 6))
    axs = axs.flatten()

    for i, (platform, sales) in enumerate(sales_per_year_pivot.iterrows()):
    
        sales = sales.drop('total_sales')
        axs[i].plot(sales.index, sales.values)
        axs[i].set_title(platform)
        axs[i].set_xlabel('Year of Release')
        axs[i].set_ylabel('Total Sales')
        axs[i].grid(True)
        
    for i in range(num_subplots, rows * 3):
        fig.delaxes(axs[i])
    plt.tight_layout()
    plt.show()

# Use the function to plot sales per year
sales_per_year(df)
```

```python
def plot_sales_per_year_boxplot(df):

    platforms = all_data['platform'].unique()
    num_platforms = len(platforms)
    num_rows = 6
    num_cols = int(np.ceil(num_platforms / num_rows))

    # Membuat subplot
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 15), sharey=True)
    axs = axs.flatten()

    for i, platform in enumerate(platforms):
        platform_data = all_data[all_data['platform'] == platform]
        sales_per_year = platform_data.groupby('year_of_release')['total_sales'].sum()
        data = [sales_per_year.values]  # Sesuaikan format data untuk boxplot
        # Plot boxplot
        axs[i].boxplot(data, labels=[platform])
        axs[i].set_title(f"{platform}\nmean: {sales_per_year.mean():.2f}")
        axs[i].set_xlabel('Total Penjualan')
        axs[i].grid(True)

    # Menghapus subplot yang tidak digunakan
    for i in range(num_platforms, num_rows * num_cols):
        fig.delaxes(axs[i])

    plt.tight_layout()
    plt.show()

# Gunakan fungsi untuk membuat boxplot
plot_sales_per_year_boxplot(df)


```

- PS2 mencatat tahun dengan penjualan tertinggi pada tahun 2004, disusul oleh Wii pada tahun 2009, dan X360 pada tahun 2010.

- Platform-platform seperti PS, PS3, dan DS juga mencatat total penjualan yang tinggi selama periode penelitian.

- Platform PS4 mencapai puncak penjualannya pada tahun 2015, menunjukkan potensi pertumbuhan yang baik bagi platform ini.

- Beberapa platform, seperti SNES dan 2600, mencapai total penjualan yang cukup tinggi meskipun sudah tidak lagi aktif pada tahun-tahun terakhir penelitian.

- PCFX adalah salah satu platform dengan penjualan terendah sepanjang periode.

## 4.3  Pengaruh Rating ESRB terhadap penjualan pada masing masing wilayah <a id='esrb_effect'></a>

```python
ESRB_dict = {'E': 'Everyone', 'M': 'Mature (> 17 years)', 'T': 'Teen (13-19 years)', 'E10+': 'Everyone 10 and older'}

# Map the ratings to descriptions and create a new column 'ESRB_description'
df['ESRB_description'] = df[df['rating'].notnull()]['rating'].map(ESRB_dict)
df.dropna(subset=['ESRB_description'], inplace=True)
```

```python
df['ESRB_description'].value_counts()
```

```python
# Group by rating and aggregate sales for each region
rating_region_agg = df.groupby('ESRB_description')[['na_sales', 'eu_sales', 'jp_sales']].agg('sum').reset_index()

def top_esrb_sales(df, region, top_n=5):
    # Sort the DataFrame based on the specified region's sales in descending order
    sorted_df = rating_region_agg.sort_values(by=f'{region}_sales', ascending=False)
    
    # Select the top N ratings for the specified region
    top_ratings = sorted_df.head(top_n)[['ESRB_description', f'{region}_sales']]
    top_ratings.reset_index(drop=True, inplace=True)
    
    return top_ratings
```

# 5. Menguji Hipotesis Statistik <a id='hypotesis'></a>

## 5.1 Rata-rata rating pengguna platform Xbox One dan PC adalah sama. <a id='avg_plat_xbox1_PC'></a>


* Pengujian Statistik: Uji T independen samples 
* Alpha (tingkat signifikansi): 0.05 (5%)
* Hipotesis:

    *  Hipotesis Nol (H0): Rata-rata rating pengguna platform Xbox One SAMA dengan rata-rata rating pengguna platform PC.
   
    *  Hipotesis Alternatif (H1): Rata-rata rating pengguna platform Xbox One BEDA dengan rata-rata rating pengguna platform PC.

```python
df['platform'].unique()
```

```python
xbox_one_ratings = df[(df['platform'] == 'XOne') & (df['user_score'] > 0)]['user_score']
pc_ratings = df[(df['platform'] == 'PC') & (df['user_score'] > 0)]['user_score']

#xbox_one_ratings = df_user_score[df_user_score['platform'] == 'XOne']['user_score'].dropna()
#pc_ratings = df_user_score[df_user_score['platform'] == 'PC']['user_score'].dropna()

print('XBox One user ratings variance:', np.var(xbox_one_ratings).round(2))
print('PC user ratings variance:', np.var(pc_ratings).round(2))
```

```python
# Uji Hipotesis 1 (Xbox One vs. PC)


alpha = 0.05

# Uji statistik (t-test)
t_statistic, p_value = st.ttest_ind(xbox_one_ratings, pc_ratings, equal_var=False)

print("Uji Hipotesis 1 (Xbox One vs. PC):")
print("T-statistic:", t_statistic)
print("P-value:", p_value)

if p_value < alpha:
    print("Hipotesis nol ditolak: Rata-rata rating pengguna platform Xbox One tidak sama dengan PC.")
else:
    print("Tidak cukup bukti untuk menolak hipotesis nol.")
```

- Berdasarkan hasil uji statistik, kita menolak hipotesis nol. Artinya, terdapat cukup bukti statistik untuk menyimpulkan bahwa rata-rata rating pengguna platform Xbox One tidak sama dengan rata-rata rating pengguna platform PC.

## 5.2 Rata-rata rating pengguna genre Action dan Sports berbeda. <a id='avg_genre_action_sports'>

* Pengujian Statistik: Uji T independen samples

* Alpha (tingkat signifikansi): 0.05 (5%)

* Hipotesis:

    * Hipotesis Nol (H0): Rata-rata rating pengguna untuk genre Action SAMA dengan rata-rata rating pengguna genre Sports.

    * Hipotesis Alternatif (H1): Rata-rata rating pengguna untuk genre Action BEDA dengan rata-rata rating pengguna genre Sports.


```python
# Uji Hipotesis 2 (Action vs. Sports)
action_ratings = df[(df['genre'] == 'Action') & (df['user_score'] > 0)]['user_score']
sports_ratings = df[(df['genre'] == 'Sports') & (df['user_score'] > 0)]['user_score']
print('XBox One user genre variance:', np.var(action_ratings).round(2))
print('PC user genre variance:', np.var(sports_ratings).round(2))
```

```python

# Uji statistik (t-test)
t_statistic, p_value = st.ttest_ind(action_ratings, sports_ratings, equal_var=False)

print("\nUji Hipotesis 2 (Action vs. Sports):")
print("T-statistic:", t_statistic)
print("P-value:", p_value)

if p_value < alpha:
    print("Hipotesis nol ditolak: Rata-rata rating pengguna genre Action tidak sama dengan Sports.")
else:
    print("Tidak cukup bukti untuk menolak hipotesis nol.")
```

- Berdasarkan hasil uji statistik, tidak terdapat cukup bukti untuk menyimpulkan bahwa rata-rata rating pengguna untuk genre Action sama dengan rata-rata rating pengguna untuk genre Sports.
